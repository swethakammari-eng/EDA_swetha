**args-kwargs**

- args means we can provide multiple value with one argument name

    - which means while calling function wea can provide directly 'virat',102

    - while define the function you no need to mention argument names also

    - the output comes in **tuple** format :('virat', 102)


- kwargs keyword arguments means we can provide multiple values with argument names

    - while define the function no need to provide the name of the arguments

    - while call the functions definitely we need to provide with argument names

    - name='virat', runs=100

    - the o/p comes in dict format: {'name' : 'virat', 'runs':102}



```python
def fun(*args):
    print(args)
    
fun(10,20)
#without name o/p in tuple format
```

```python
def fun2(**kwargs):
    print(kwargs) # {'a':10,'b':20}

fun2(a=10,b=20)
#with name o/p in dict format
```
