```python
- any loop we require 3 things

- initial point(start point)

- increment or decrement

- condition to stop

- in for loop the above  are happen in a single line

- but in while loop all 3 needs 3 different lines

```

```python
#start point
#while <condition>:
#start your code
#increment or decrement

range(1,10,1)
#start=1 increment=1 stop=10-1=9

i=1
while <condition>:
    i=i+1
```
