**try-except**

- try except is used to handle the exeption mechanism

- try except is used to catch and handle the errors

- when we get the error python will stop the execution

- using try except eventhough we get the errors, the execution will not stop

- it will catch all the logical errors automatically

- the actual code block will write in **try block**

- if any error occurs in try block, will catch the errors in **except** block

- syntax errors will not catch



    - catch all errors

    - code execution will not stop

    - original is in try block

    - error handling== except block

**try-except-finally**

- original code in **try** block

- error catch in **except** block

- **finally** block will activate either  or try and except activate

- assume that try block executed then finally block can also execute

- assume that except block executed then also finally block executed

- even though attempt is success or fail we can execute finally(coffee ex: good coffee try, bad coffee exception, kitchen should clean  finally)

```python
try:
    print('connection established')
except Exception as e:
    print(e)
finally:
    print('connection closed')
```

```python
try:
    100/0
    print('hi')
except Exception as e:
    print(e)
    print('error aagaya isliye mai aagaya')
finally:
    print('connection closed')
```
