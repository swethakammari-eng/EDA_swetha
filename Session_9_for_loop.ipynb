### **loops**

- loops are used to execute a bloack of code multiple times

- 3 parameters

    - start point

    - increment or decrement

    - condition to stop

- for loop and while loop

- if we know the condition when to stop **for loop** prefer

- if we dont know when to stop **while** loop prefer

**case-1: range(stop)**

- start point by default=0

- direction: positive direction

- positive direction  means **increment**

- by default increment by 1

- last point=stop-1

**case-3: range(start,stop,step)**

- start value= start

- if step is positive sign

    - it is positive direction (increment)

    - increment by **step value**

    - last value= stop-1

- if step is negative sign

    - it is negative direction (decrement)

    - decrement by **step value**

    - last value= stop+1

```python
range(1,20,2)
#start=1
#step= +2
#increment by 2
#last=stop-1 =20-2=18
```

```python
for i in range(1,20,2):
    print(i, end=' ')
```

**Continue-pass-break**

```python
for i in range(5):
    if i==2:
        print('great')
    print(i)

#s-1: i=0 if 0==2 F print(0)
#s-2: i=1 if 1==2 F print(1)
#s-3: i=2 if 2==2 T great print(2)
#s-4: i=3 if 3==2 F print(3)
#s-4: i=4 if 4==2 F print(4)
# 0 1 great 2 3 4
```

```python
for i in range(5):
    if i==2:
        print('great') #here we missed 2
        continue
    print(i, end='===>')
```

```python
for i in range(5):
    if i==2:
        continue # here we missed great
        print('great')
    print(i, end='==>')
```

```python
for i in range(5):
    if i==2:
        print('great') # here we missed super
        continue
        print('super')
    print(i, end='===>')
print('okay')
```

```python
import random
even_count=0
odd_count=0
for i in range(5):
    num=random.randint(1,100)
    if num%2==0:
        even_count=even_count+1
        print(f'{num} is even')
    else:
        odd_count=odd_count+1
        print(f'{num} is odd')
print(f'the total even count is: {even_count}')
print(f'the total odd count is: {odd_count}')
```

```python
num=eval(input('enter a number'))
even_count=0
odd_count=0
even_summ=0
odd_summ=0
for i in range(1,num+1):
    if num%i==0:
        even_count=even_count+1
        even_summ=even_summ+num
    else:
        odd_count=odd_count+1
        odd_summ=odd_summ+num
        
print(f'total even count is: {even_count}')
print(f'total odd count is: {odd_count}')
print(f'sum of even numbers is: {even_summ}')
print(f'sum of odd numbers is: {odd_summ}')        
```
